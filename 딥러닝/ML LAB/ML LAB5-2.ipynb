```python
#Lab 5 Logistic Regression Classifier

import tensorflow as tf
import numpy as np

xy = np.loadtxt('data-03-diabetes.csv', delimiter = ',', dtype = np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:,[-1]]
```

```python
print(x_data.shape, y_data.shape)
```

```python
#target data 에는 0과1 밖에 없는 걸 확인 --> sigmoid 함수 사용 가능
print(np.unique(y_data, return_counts = True))
```

심층 신경망 만들기 위한 tf.keras.Sequential() 호출

```python
tf.model= tf.keras.Sequential()
```

tf.mode.add(tf.keras.layers.Dense()) 를 이용하여 층 추가하기

parameter로는 units, input_dim(입력층 개수 == x_data.shape[1]), 활성함수

```python
#multi-variable, x_data.shape[1] == faeture counts == 8 in this case
tf.model.add(tf.keras.layers.Dense(units = 1, input_dim = x_data.shape[1], activation = 'sigmoid'))
```

compile 메서드를 이용

parameter로는 오차함수(loss function), optimizer(SGD <- loss 의 min값 찾는 방법), metrics

```python
import warnings
warnings.filterwarnings('ignore')

tf.model.compile(loss = 'binary_crossentropy', optimizer = tf.keras.optimizers.SGD(lr = 0.01), metrics = ['accuracy'])
```

```python
tf.model.summary()

#첫번째 Param이 9 인 이유는 8*1(1개의 뉴런) + 1(b)
```

```python
history = tf.model.fit(x_data, y_data, epochs = 500)
```

```python
#accuracy 마지막 epoch꺼 출력하기 : [-1]
print('Accuracy: {0}'.format(history.history['accuracy'][-1]))
```

```python
#predict a single data point

y_predict = tf.model.predict([[0.176471, 0.155779, 0, 0, 0, 0.052161, -0.952178, -0.733333]])
```

```python
print('Predict: {0}'.format(y_predict))
```

```python
#evaluating model
#evaluate의 [0] 번째는 cost function 의 값, [1]번째는 accuracy 값

evaluate = tf.model.evaluate(x_data,y_data)
print('loss : {0}, accuracy : {1}'.format(evaluate[0],evaluate[1]))
```
