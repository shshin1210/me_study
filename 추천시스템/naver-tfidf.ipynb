```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

path = 'C:/Users/queenSSH/Desktop/학교/CUAI/cuai하계/네이버 크롤링/'
```

```python
naver = pd.read_csv(path + '네이버_추천시스템.csv' , encoding='cp949')
```

```python
naver.head()
```

```python
naver = naver.drop('Unnamed: 0', axis =1)
```

```python
naver = naver[['title', 'genre', 'story']]
```

```python
tfidf = TfidfVectorizer()

#story에 대해서 tf-idf 수행
tfidf_matrix = tfidf.fit_transform(naver['story'])
print(tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_matrix.shape
```

```python
np.round(cosine_matrix,4)
```

```python
#movie title와 id를 매핑할 dictionary 생성

naver2id = {}
for i, c in enumerate(naver['title']): 
    naver2id[i] = c


#id와 movie 샤싣dmf 매핑할 dictionary 생성
id2naver = {}
for i,c in naver2id.items() : 
    id2naver[c] = i
```

```python
print(id2naver)
```

```python
title_list = naver['title']
```

```python
print(title_list[0])
```

```python
print(len(title_list))
```

```python
print(naver['title'][0])
```

```python
#마물전담부대의 id 추출

for j in range(len(title_list)):
    #idx = id2naver['마물전담부대'] #마물전담부대 :  0번 인덱스
    idx = id2naver[title_list[j]]
    print(naver2id[j])
    sim_scores = [(i,c) for i,c in enumerate(cosine_matrix[idx]) if i != idx]
    #자기 자신을 제외한 영화들의 유사도 및 인덱스 추출

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse = True) #유사도가 높은 순서대로 정렬
    sim_scores[0:10] #상위 10개의 인덱스와 유사도 추출
    
    sim_scores = [(naver2id[i], score) for i, score in sim_scores[0:10]]
    print(sim_scores)
```
