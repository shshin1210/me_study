```python
import tensorflow as tf
import numpy as np

sentence = ("if you want to build a ship, don't drum up people together to "
            "collect wood and don't assign them tasks and work, but rather "
            "teach them to long for the endless immensity of the sea.")
```

### 데이터 전처리

```python
char_set = list(set(sentence))
char_dic = {w : i for i,w in enumerate(char_set)} # w 가 word, i 가 index
```

```python
print(char_set)
```

```python
print(char_dic)
```

```python
data_dim = len(char_set)
hidden_size = len(char_set)
num_classes = len(char_set)
sequence_length = 10 # Any arbitrary number
learning_rate = 0.1
```

```python
dataX =[]
dataY =[]

for i in range(0, len(sentence) - sequence_length):
    x_str = sentence[i:i+sequence_length]
    y_str = sentence[i+1:i+sequence_length+1]
    
    print(i, x_str, '->', y_str)
    
    x = [char_dic[c] for c in x_str] # x_str to index
    y = [char_dic[c] for c in y_str] # y_str to index
     
    dataX.append(x) #index 대입
    dataY.append(y)
```

```python
print(len(dataX))
print(dataX[0])
```

```python
batch_size = len(dataX)
```

```python
# One -hot Encoding

X_one_hot = tf.one_hot(dataX, num_classes)
Y_one_hot = tf.one_hot(dataY, num_classes)
```

```python
print(X_one_hot)
```

```python
print(X_one_hot.shape)  # check out the shape (170, 10, 25)
print(Y_one_hot.shape)  # check out the shape
```

### Model 생성

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.LSTM(units = num_classes, input_shape=(sequence_length, X_one_hot.shape[2]),
                                 return_sequences = True))
tf.model.add(tf.keras.layers.LSTM(units = num_classes,return_sequences = True))
tf.model.add(tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(units = num_classes,
                                                                  activation = 'softmax')))
tf.model.summary()
```

```python
tf.model.compile(loss = 'categorical_crossentropy', optimizer = tf.keras.optimizers.Adam(lr=learning_rate),
                metrics=['accuracy'])
```

```python
tf.model.fit(X_one_hot, Y_one_hot, epochs = 100)
```

# enumerate 부분.....

```python
results = tf.model.predict(X_one_hot)

for j, results in enumerate(results):
    index = np.argmax(results, axis =1)

    if j==0: #print all for the first result to make a sentenxe
        print(''.join([char_set[t] for t in index]), end ='')
    else:
        print(char_set[index[-1]], end='')
```
